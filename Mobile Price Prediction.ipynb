```python
model= RandomForestClassifier()
model2= SVC()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df1.drop('price_range',axis=1), df1['price_range'], test_size=0.33, random_state=42)
```

```python
cross_val_score(model,X_test,y_test,cv=10).mean()
```

```python
cross_val_score(model2,X_test,y_test,cv=10).mean()
```

```python
model.fit(X_train,y_train)
```

```python
model2.fit(X_train,y_train)
```

```python
pred=model.predict(X_test)
```

```python
print(classification_report(y_test,pred))
```

```python
pred2=model2.predict(X_test)
```

```python
print(classification_report(y_test,pred2))
```

```python
model3= DecisionTreeClassifier()
```

```python
cross_val_score(model3,X_test,y_test,cv=10).mean()
```

```python
model3.fit(X_train,y_train)
```

```python
print(classification_report(y_test,model3.predict(X_test)))
```
